importing all the libraries we need to start the work

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
df = pd.read_csv('AmesHousing.csv')
```

same code in the TP to find the missing values and hot encode the categorical values

```python
#Valeurs manquantes
VM = pd.DataFrame({
'Colonne': df.columns.values,
'nbr de VM': df.isna().sum().values,
'% de VM': 100 * df.isna().sum().values / len(df),})
VM = VM[VM['nbr de VM'] > 0]
print(VM.sort_values(by='nbr de VM',
ascending=False).reset_index(drop=True))
```

```python
#on affiche les variables catГ©gorielles et les variables numГ©riques
colonnes_avec_VM = df.columns[df.isna().sum() > 0]
for col in colonnes_avec_VM :
   print(col)
   print(df[col].unique()[:5])
   print('*'*30)
```

```python
#on remplace les VM numГ©riques par la moyenne de la colonne
num_VM = ['Lot Frontage', 'Mas Vnr Area', 'BsmtFin SF 1', 'BsmtFin SF 2',
'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Full Bath',
'Bsmt Half Bath', 'Garage Yr Blt', 'Garage Cars', 'Garage Area']
for n_col in num_VM:
    df[n_col] = df[n_col].fillna(df[n_col].mean())
```

```python
#On remplace les VM nominales par le mode variable
nom_VM = [x for x in colonnes_avec_VM if x not in num_VM]
for nom_col in nom_VM:
    df[nom_col] = df[nom_col].fillna(df[nom_col].mode().to_numpy()[0])
```

```python
#encodage des variables catГ©gorielles
#on affiche tout d'abord les types de donnГ©es pour chaque colonne
types = pd.DataFrame({
'Colonne': df.select_dtypes(exclude='object').columns.values,
'Type': df.select_dtypes(exclude='object').dtypes.values})
print(types)
```

```python
#encodage des variables catГ©gorielles,
#selon le descriptif de la dataset la colonne MS SubClass est nominale,
#Pandas a du mal Г  retourner son vrai type
df['MS SubClass'] = df['MS SubClass'].astype(str)

selection_de_variables = ['MS SubClass', 'MS Zoning', 'Lot Frontage',
'Lot Area',

'Neighborhood', 'Overall Qual', 'Overall Cond',
'Year Built', 'Total Bsmt SF', '1st Flr SF', '2nd Flr SF',
'Gr Liv Area', 'Full Bath', 'Half Bath', 'Bedroom AbvGr',
'Kitchen AbvGr', 'TotRms AbvGrd', 'Garage Area',
'Pool Area', 'SalePrice']
df = df[selection_de_variables]
#la dataset comprendera maintenant que 67 variables
df = pd.get_dummies(df)
```

split data on train and test data

```python
#fractionner dataset en des donnГ©es de test et de train
train = df.sample(frac = 0.8, random_state = 9)
test = df.drop(train.index)
#variable cible
train_cible = train.pop('SalePrice')
test_cible = test.pop('SalePrice')
```

data standarisation

```python
#Standardisation
variables_pred = train.columns
for col in variables_pred:
   col_moyenne = train[col].mean()
   col_ecart_type = train[col].std()
   if col_ecart_type == 0:
      col_ecart_type = 1e-20
train[col] = (train[col] - col_moyenne) / col_ecart_type
test[col] = (test[col] - col_moyenne) / col_ecart_type
#Construction du modГЁle dвЂ™apprentissage profond


modele = keras.Sequential([
layers.Dense(64, activation = 'relu', input_shape =
[train.shape[1]]),
layers.Dropout(0.3, seed = 2),
layers.Dense(64, activation = 'swish'),
layers.Dense(64, activation = 'relu'),
layers.Dense(64, activation = 'swish'),
layers.Dense(64, activation = 'relu'),
layers.Dense(64, activation = 'swish'),
layers.Dense(1)])
```

change the optimizer to adam

```python
optimiseur = tf.keras.optimizers.Adam(learning_rate = 0.0005)
modele.compile(loss = tf.keras.losses.MeanSquaredError(),optimizer =
optimiseur,metrics = ['mae'])
```

here ive change to earlystopping because the model was not accurate at all so i've tried to add also the epochs number and it work give me a close number but higher than the price

```python

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_mae', # Monitor the validation Mean Absolute Error
    patience=10,       # Number of epochs with no improvement after which training will be stopped
    restore_best_weights=True # Restore model weights from the epoch with the best value of the monitored quantity
)

training = modele.fit(train, train_cible, epochs = 100, # Increase epochs, EarlyStopping will handle when to stop
                        validation_split = 0.2,
                        callbacks=[early_stopping])
```

```python
historique = pd.DataFrame(training.history)
historique['epoque'] = training.epoch
figure, axe = plt.subplots(figsize = (14,8))
num_epoque = historique.shape[0]
axe.plot(np.arange(0, num_epoque), historique["mae"],label = "TrainingMAE", lw = 3, color = 'red')
axe.plot(np.arange(0, num_epoque), historique["val_mae"],label =
"Validation MAE", lw = 3, color = 'blue')
axe.legend ()
plt.tight_layout()
plt.show()
```

there is no sign of overfitting from the plot we got.as i know

```python
test1 = test.iloc[[10]]
test_prediction = modele.predict(test1).squeeze()
test_label = test_cible.iloc[10]
print("PrГ©diction du modГЁle = {:.2f}".format(test_prediction))
print("Valeur actuelle = {:.2f}".format(test_label))
print("Erreur absolue = {:.2f}".format(abs(test_prediction - test_label)))
print('accuracy'.format(100 * abs(test_prediction - test_label) / test_label))
print(format(100 * abs(test_prediction - test_label) / test_label))
```
